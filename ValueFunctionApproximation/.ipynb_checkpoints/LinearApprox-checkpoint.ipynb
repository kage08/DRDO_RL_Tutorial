## Task
In this tutorial we use value function approximation to solve a popular reinforcement learning task: [MountainCar](https://gym.openai.com/envs/MountainCar-v0/) using Q-learning.

![Mountain Car Environment](./mountaincar.png)

The task is to drive an under-powered car up the hill. The best strategy is to build momentum by going left and right alternatively till we reach the flag after building sufficient kinetic energy.

```python
# Import libraries
import gym
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
from tqdm import tqdm
%matplotlib inline
```

```python
env = gym.make("MountainCar-v0")
env.reset()
print("Bound on state:",env.observation_space.high, env.observation_space.low)
print("Action Space:", env.action_space)
```

The state is array: $[position, velocity]$. 

The actions are discrete: $[left,right,neutral]$.

We can't (directly) use tabular methods since the state space is continuous. So we use a linear function approximator to model $Q(s,a)$: for each action $a$, $$Q(s,a)=f_a(s)=\sum_i w_i s_i$$
where $s_i$ is the ith feature of state.

We use RBFSampler to approximate feature map of RBF kernel to get features.

```python

from sklearn.kernel_approximation import RBFSampler

import sklearn.pipeline
import sklearn.preprocessing

observation_examples = np.array([env.observation_space.sample() for x in range(10000)])
scaler = sklearn.preprocessing.StandardScaler()
scaler.fit(observation_examples)
featurizer = sklearn.pipeline.FeatureUnion([
        ("rbf1", RBFSampler(gamma=5.0, n_components=100)),
        ("rbf2", RBFSampler(gamma=2.0, n_components=100)),
        ("rbf3", RBFSampler(gamma=1.0, n_components=100)),
        ("rbf4", RBFSampler(gamma=0.5, n_components=100))
        ])
featurizer.fit(scaler.transform(observation_examples))

def get_features(state):
    return featurizer.transform(scaler.transform([state]))[0]
```

```python
#Stochatic Gradient Updater for f_a(s)
from sklearn.linear_model import SGDRegressor
lr = 0.01

action_models = []
for _ in range(env.action_space.n):
    model = SGDRegressor(learning_rate="constant", eta0=lr)
    model.partial_fit([get_features(env.reset())],[0.0])
    action_models.append(model)

def predict_Q(state):
    return np.array([model.predict([get_features(state)])[0] for model in action_models])
```

Now we define function to update weights. SGDRegressor automatically does Stochastic Gradient descent update of weights given the TD-estimate.

```python
def update_weights(state, action, target):
    action_models[action].partial_fit([get_features(state)], [target])
```

Now we are set to perform Q-learning. First lets define our exploration strategy/ies and hyperparameters.

```python
from scipy.special import softmax
seed = 2
rg = np.random.RandomState(seed)

# Epsilon greedy
def choose_action_epsilon(state, epsilon, rg=rg):
    if rg.rand() < epsilon:
        return env.action_space.sample()
    else:
        return predict_Q(state).argmax()

#Softmax
def choose_action_softmax(state, rg=rg):
    return rg.choice(env.action_space.n, p = softmax(predict_Q(state).ravel()))
```

```python
gamma = 1.0
episodes = 1000
epsilon0 = 0.0
epsilon_decay = 1.0
```

Finally the Q-learning algorithm

```python
def q_learn(env, choose_action=choose_action_epsilon, num_episodes=episodes, render=False):
    episode_rewards = np.zeros(episodes)
    steps_to_completion = np.zeros(episodes)
    epsilon = epsilon0
    
    for i_episode in range(num_episodes):
        
        # Reset the environment and pick the first action
        epsilon *= epsilon_decay
        state = env.reset()
        done = False
        tot_reward, steps = 0, 0
        
        
        # One step in the environment
        while not done:
                        
            action = choose_action(state, epsilon)
            
            next_state, reward, done, _ = env.step(action)
            
            td_target = reward + gamma * predict_Q(next_state).max()
            update_weights(state, action, td_target)
            
            tot_reward += reward
            steps += 1        
                
            state = next_state
            if render:
                env.render()
            
        episode_rewards[i_episode] = tot_reward
        steps_to_completion[i_episode] = steps
        print("\rEpisode {}/{} Steps:{}".format(i_episode+1,num_episodes,steps), end="")
    
    return episode_rewards, steps_to_completion
    
```

```python
stats = q_learn(env, num_episodes=1000)
```

```python
plt.figure()
plt.xlabel("Episodes")
plt.ylabel("No. of steps")
plt.plot(np.arange(len(stats[1]))+1, stats[1])
```

```python
plt.figure()
plt.xlabel("Episodes")
plt.ylabel("Reward")
plt.plot(np.arange(len(stats[0]))+1, stats[0])
```

Let's see the value function learnt

```python
v_range = np.linspace(-0.07, 0.07, 100)
p_range = np.linspace(-1.2, 0.6, 100)
x_range, y_range = np.meshgrid(v_range, p_range)
def val(v,p):
    return predict_Q([v,p]).max()
    
v = np.vectorize(val)(x_range, y_range)
```

```python
from mpl_toolkits.mplot3d import Axes3D
plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(x_range, y_range, v, 50)
ax.set_xlabel('Velocity')
ax.set_ylabel('Position')
ax.set_zlabel('Value')
```
